```python
from sklearn.ensemble import IsolationForest
import pandas as pd
import numpy as np
```

```python
train_data = pd.read_excel("train.xlsx")
```

```python
train_data.head()
```

```python
train_data = train_data[['tarih', 'id', 'gider']]
```

```python
train_data.head()
```

```python
train_data['tarih'] = train_data['tarih'].values.astype('float64')
```

```python
train_data.head()
```

```python
model = IsolationForest(contamination=0.1)
```

```python
model.fit(train_data)
```
